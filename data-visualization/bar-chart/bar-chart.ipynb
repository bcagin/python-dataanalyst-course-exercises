```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_used_car = pd.read_csv('bar_chart_data.csv')
```

```python
df_used_car
```

```python
plt.figure(figsize = (9,6))
plt.bar(x = df_used_car['Brand'],
        height = df_used_car['Cars Listings'],
        color = 'midnightblue')
plt.xticks(rotation = 45, fontsize = 13)
plt.yticks(fontsize = 13)
plt.title('Car Listings by Brand', fontsize = 16, fontweight = 'bold')
plt.ylabel('Number of Listings', fontsize = 13)
plt.savefig("ChartCars.png", bbox_inches = "tight")
plt.show()
```
