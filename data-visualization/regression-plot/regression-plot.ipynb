```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_reg = pd.read_csv('scatter_plot_ii.csv')
```

```python
df_reg
```

```python
sns.set(rc = {'figure.figsize' : (9,6)})
sns.regplot(x = 'Budget',
           y = 'Sales',
           data = df_reg,
           scatter_kws = {'color' : 'k'},
           line_kws = {'color' : 'red'})
plt.xlabel('Ad Expenditure in (000s $)')
plt.ylabel('Sales in (000s in Units)')
plt.title('Effect of Ad Expenditure on Sales', fontsize = 14, fontweight = 'bold')
plt.savefig('Effect of Ad Expenditure on Sales')
plt.show()
```
